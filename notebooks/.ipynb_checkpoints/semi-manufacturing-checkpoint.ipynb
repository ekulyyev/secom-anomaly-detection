# Unsupervised Anomaly Detection on Semiconductor Manufacturing 
Semiconductor manufacturing processes generate high-dimensional sensor data that can indicate equipment faults or process deviations. Labeling failures is costly and time-consuming; therefore, unsupervised methods that learn normal behavior and flag deviations are highly valuable. This project focuses on:
- Building a robust pipeline for raw sensor data ingestion and preprocessing.
- Comparing several unsupervised algorithms for anomaly detection.
- Selecting and tuning a final model that optimally balances recall on true failures and false-alarm rate.
- Packaging the solution into a deployable pipeline.

```python
import os
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, roc_auc_score, recall_score
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.neighbors import LocalOutlierFactor
from sklearn.pipeline import Pipeline
import joblib



# paths
RAW_DIR = os.path.join("..", "data", "raw")
PROCESSED_DIR = os.path.join("..", "data", "processed")

print("Environment ready. Data dir:", RAW_DIR)
```

## Load Raw SECOM Data
Read the secom.data and display its shape and initial rows.

```python
# load feature matrix
secom_path = os.path.join(RAW_DIR, "secom.data")
X = pd.read_csv(secom_path, sep="\s+", header=None)
print("Features shape:", X.shape)

# load labels (for evaluation only)
labels_path = os.path.join(RAW_DIR, "secom_labels.data")
y_raw = pd.read_csv(labels_path, sep="\s+", header=None, names=["index","label"])
y = y_raw.set_index("index")["label"]
print("Labels distribution:\n", y.value_counts())

# preview as always
display(X.head())
```

```python
# load labels
labels_df = pd.read_csv(
    labels_path,
    sep=r"\s+",
    header=None,
    names=["flag", "date", "time"]
)

# combine date + time into a single datetime
labels_df["datetime"] = pd.to_datetime(
    labels_df["date"] + " " + labels_df["time"],
    dayfirst=True,           # file uses dd/mm/YYYY
    format="%d/%m/%Y %H:%M:%S"
)

# convert the “flag” (.1 = pass, 1 = fail) into a 0/1 label
y = (labels_df["flag"] == 1.0).astype(int)
print("Label distribution:\n", y.value_counts())

# align index with X
y.index = X.index
```

## Missing-Value Analysis & Imputation
Drop high-NaN sensors and impute remaining missing values.

```python
# compute missingness per feature
missing_pct = X.isnull().mean().sort_values(ascending=False)
print("Top 10 features by % missing:")
print((missing_pct.head(10) * 100).round(2).astype(str) + "%\n")

# drop features with >50% missing values
threshold = 0.50
to_drop = missing_pct[missing_pct > threshold].index
X_reduced = X.drop(columns=to_drop)
print(f"Dropped {len(to_drop)} features ({round(len(to_drop)/X.shape[1]*100,1)}% of total) with >{threshold*100:.0f}% missing.\n")

# impute remaining missing values
imp = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(
    imp.fit_transform(X_reduced),
    columns=X_reduced.columns,
    index=X_reduced.index
)

print("After imputation, total missing values:", X_imputed.isnull().sum().sum())

# save processed feature matrix
os.makedirs(PROCESSED_DIR, exist_ok=True)
X_imputed.to_csv(os.path.join(PROCESSED_DIR, "secom_X_imputed.csv"), index=False)
print(f"Imputed data saved to {os.path.join(PROCESSED_DIR, 'secom_X_imputed.csv')}")
```

## PCA for Dimensionality Reduction
Fit PCA to 50 components and plot explained‐variance to compress data.

```python
# run PCA to reduce to 50 components
n_components = 50
pca = PCA(n_components=n_components, random_state=42)
X_pca = pca.fit_transform(X_imputed)

# examine explained variance
explained_ratio = pca.explained_variance_ratio_
cum_explained = np.cumsum(explained_ratio)

print("Explained variance ratio (first 10 PCs):")
print(np.round(explained_ratio[:10], 4))
print("\nCumulative explained variance (first 10 PCs):")
print(np.round(cum_explained[:10], 4))
print(f"\nTotal variance captured by {n_components} PCs: {cum_explained[-1]:.4f}")

# plot cumulative explained variance
plt.figure(figsize=(8, 4))
plt.plot(range(1, n_components+1), cum_explained, marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance Curve")
plt.grid(True)
plt.tight_layout()
plt.show()

# save the reduced dataset
X_pca_df = pd.DataFrame(
    X_pca,
    index=X_imputed.index,
    columns=[f"PC{i+1}" for i in range(n_components)]
)
X_pca_df.to_csv(os.path.join(PROCESSED_DIR, "secom_X_pca50.csv"), index=False)
print(f"PCA-transformed data saved to {os.path.join(PROCESSED_DIR, 'secom_X_pca50.csv')}")
```

## Isolation Forest Training & Evaluation
Train IF on PCA data, evaluate with classification report and ROC AUC.

```python
# fit Isolation Forest on the PCA‐reduced data
# prepare
iso = IsolationForest(
    n_estimators=100,
    contamination=0.1,      # rough guess (10% anomalies)
    max_samples='auto',
    random_state=42
)
iso.fit(X_pca)

# compute anomaly scores (higher → more anomalous)
scores = -iso.decision_function(X_pca)

# choose threshold and label anomalies
# here we’ll use the 90th percentile of scores as threshold,
# but you can tune this based on your desired false‐alarm rate.
import numpy as np
threshold = np.percentile(scores, 90)
y_pred = (scores >= threshold).astype(int)

# evaluation against true labels
print("Classification report (0 = normal, 1 = anomaly):\n")
print(classification_report(y, y_pred, target_names=["pass","fail"]))

print("ROC AUC:", round(roc_auc_score(y, scores), 4))

# score distribution plot
plt.figure(figsize=(8,4))
plt.hist(scores[y==0], bins=50, alpha=0.6, label="pass (0)")
plt.hist(scores[y==1], bins=50, alpha=0.6, label="fail (1)")
plt.axvline(threshold, color="red", linestyle="--", label=f"Threshold ({threshold:.2f})")
plt.title("Isolation Forest Anomaly Score Distribution")
plt.xlabel("Anomaly Score")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

## Built-in IF threshold

```python
# built-in IF threshold
y_pred_if = (iso.predict(X_pca) == -1).astype(int)   # -1 = anomaly → 1
print(classification_report(y, y_pred_if, target_names=["pass","fail"]))
```

## Local Outlier Factor

```python
# local outlier factor
lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=158/1567,   # match the true fail rate (~10.1%)
)
y_pred_lof = (lof.fit_predict(X_pca) == -1).astype(int)
scores_lof = -lof.negative_outlier_factor_

print("=== LOF Classification ===")
print(classification_report(y, y_pred_lof, target_names=["pass","fail"]))
print("ROC AUC (LOF):", round(roc_auc_score(y, scores_lof), 4))
```

## IF directly on X_imputed

```python
# IF directly on X_imputed
iso_full = IsolationForest(contamination=y.mean(), random_state=42)
iso_full.fit(X_imputed)
scores_full = -iso_full.decision_function(X_imputed)
y_pred_full = (iso_full.predict(X_imputed) == -1).astype(int)

print(classification_report(y, y_pred_full, target_names=["pass","fail"]))
print("ROC AUC:", roc_auc_score(y, scores_full))
```

## Contamination Sweep for IsolationForest

```python
# find the best recall vs. false-alarm trade-off
cont_levels = np.linspace(0.01, 0.20, 20)        # from 1% up to 20% anomalies
results = []

for c in cont_levels:
    iso = IsolationForest(contamination=c, random_state=42)
    iso.fit(X_imputed)
    y_pred = (iso.predict(X_imputed) == -1).astype(int)
    
    recall_fail = recall_score(y, y_pred, pos_label=1)
    fp_rate   = ( (y_pred == 1) & (y == 0) ).sum() / (y == 0).sum()
    results.append((c, recall_fail, fp_rate))

res_df = pd.DataFrame(results, columns=["contamination","recall_fail","false_positive_rate"])
display(res_df)

# plot recall vs. false-positive rate
plt.figure(figsize=(8,4))
plt.plot(res_df["contamination"], res_df["recall_fail"], marker="o", label="Recall (fail)")
plt.plot(res_df["contamination"], res_df["false_positive_rate"], marker="s", label="False-alarm rate")
plt.xlabel("contamination parameter")
plt.ylabel("Rate")
plt.title("IsolationForest Performance vs. contamination")
plt.legend()
plt.tight_layout()
plt.show()
```

## Cell 11

```python
# Cell 11: Final IF with contamination=0.10 and anomaly inspection
iso_final = IsolationForest(contamination=0.10, random_state=42)
iso_final.fit(X_imputed)

# Flags: -1 → anomaly (fail), +1 → normal (pass)
flags = iso_final.predict(X_imputed)
y_pred_final = (flags == -1).astype(int)

# Confirm performance
from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y, y_pred_final, target_names=["pass","fail"]))
print("ROC AUC:", round(roc_auc_score(y, -iso_final.decision_function(X_imputed)),4))

# List out the first 10 anomalies
anomaly_indices = X_imputed.index[y_pred_final == 1]
print("First 10 flagged anomaly indices:", anomaly_indices[:10].tolist())

# (Optional) view their PCA coords to see clustering
print("\nTheir PC1 vs. PC2 coordinates:")
print(X_pca_df.loc[anomaly_indices[:10], ["PC1","PC2"]])
```

## Cell 12

```python

# 1) Copy PC1/PC2 into a new DataFrame
plot_df = X_pca_df[["PC1", "PC2"]].copy()

# 2) Assign predicted_fail directly from the NumPy array 
#    (pandas will broadcast it row‐wise since shapes match)
plot_df["predicted_fail"] = y_pred_final

# 3) Assign true_fail directly from your pandas Series y
plot_df["true_fail"] = y

# 4) Plot
plt.figure(figsize=(8,6))
# normal points
plt.scatter(
    plot_df.loc[plot_df["predicted_fail"] == 0, "PC1"],
    plot_df.loc[plot_df["predicted_fail"] == 0, "PC2"],
    s=20, alpha=0.3, color="green", label="predicted pass"
)
# anomalies
plt.scatter(
    plot_df.loc[plot_df["predicted_fail"] == 1, "PC1"],
    plot_df.loc[plot_df["predicted_fail"] == 1, "PC2"],
    s=30, alpha=0.8, marker="x", color="red", label="predicted fail"
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Isolation Forest: Predicted Anomalies in PCA Space")
plt.legend()
plt.tight_layout()
plt.show()
```

## Cell 13

```python
# build a loadings DataFrame: each row = original feature, each column = PC
loadings = pd.DataFrame(
    pca.components_.T,
    index=X_imputed.columns,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)]
)

# look at the top contributors for PC1 and PC2
top_n = 10
print(f"Top {top_n} features by absolute loading on PC1:")
print(loadings["PC1"].abs().sort_values(ascending=False).head(top_n))
print("\nTop {top_n} features by absolute loading on PC2:")
print(loadings["PC2"].abs().sort_values(ascending=False).head(top_n))

# create a combined table for side-by-side comparison
top1 = loadings["PC1"].abs().sort_values(ascending=False).head(top_n).index
top2 = loadings["PC2"].abs().sort_values(ascending=False).head(top_n).index
comparison = pd.DataFrame({
    "PC1_loading": loadings.loc[top1, "PC1"],
    "PC2_loading": loadings.loc[top2, "PC2"]
})
print("\nComparison table:")
print(comparison)

# visualize as a bar chart
plt.figure(figsize=(10, 4))
loadings["PC1"].abs().sort_values(ascending=False).head(top_n).plot.bar()
plt.title("Top 10 Feature Loadings on PC1")
plt.ylabel("Absolute Loading")
plt.tight_layout()
plt.show()
```

## Cell 14

```python
# compare raw sensor readings for top-loading features
top_feats = [162, 161, 297, 296]
fig, axes = plt.subplots(len(top_feats), 1, figsize=(8, 3*len(top_feats)))

for ax, feat in zip(axes, top_feats):
    ax.hist(
        X_imputed.loc[y_pred_final==0, feat],
        bins=50, alpha=0.6, color="green",label="pred pass", density=True
    )
    ax.hist(
        X_imputed.loc[y_pred_final==1, feat],
        bins=50, alpha=0.6, color="orange", label="pred fail", density=True
    )
    ax.set_title(f"Sensor {feat} distribution")
    ax.legend()

plt.tight_layout()
plt.show()
```

## Cell 15

```python
# simple threshold rule on sensor 162
feat = 162

# choose threshold at the 10th percentile of the healthy (pass) distribution:
# flag the bottom 10% of pass samples as anomalies
threshold = np.percentile(
    X_imputed.loc[y == 0, feat],
    10
)
print(f"Using threshold {threshold:.1f} on sensor {feat}")

# predict anomaly if sensor reading < threshold
y_pred_thresh = (X_imputed[feat] < threshold).astype(int)

# evaluate
print(classification_report(y, y_pred_thresh, target_names=["pass","fail"]))
```

## Cell 16

```python
# two-sensor rule vs small IsolationForest
# features
low_feat, high_feat = 162, 161

# thresholds from the healthy (pass) distribution
low_thr  = np.percentile(X_imputed.loc[y==0, low_feat], 10)   # bottom 10%
high_thr = np.percentile(X_imputed.loc[y==0, high_feat], 90)  # top 10%
print(f"Thresholds → sensor {low_feat} < {low_thr:.1f}, sensor {high_feat} > {high_thr:.1f}\n")

# 2d threshold rules
cond_low  = X_imputed[low_feat]  < low_thr
cond_high = X_imputed[high_feat] > high_thr

y_pred_and = (cond_low & cond_high).astype(int)
y_pred_or  = (cond_low | cond_high).astype(int)

print("=== AND rule (both conditions) ===")
print(classification_report(y, y_pred_and, target_names=["pass","fail"]))

print("=== OR rule (either condition) ===")
print(classification_report(y, y_pred_or,  target_names=["pass","fail"]))


# mini-IsolationForest on the two most important sensors
X_two = X_imputed[[low_feat, high_feat]]
iso2 = IsolationForest(contamination=0.10, random_state=42)
iso2.fit(X_two)
y_pred_if2 = (iso2.predict(X_two) == -1).astype(int)

print("=== IF on [162,161] ===")
print(classification_report(y, y_pred_if2, target_names=["pass","fail"]))
```

## Cell 17

```python
# Cell 17: Mini-IF on top 4 PCA-driving sensors
top4 = [162, 161, 297, 296]
X_top4 = X_imputed[top4]

iso4 = IsolationForest(contamination=0.10, random_state=42)
iso4.fit(X_top4)
y_pred4 = (iso4.predict(X_top4) == -1).astype(int)
scores4 = -iso4.decision_function(X_top4)

print("=== IF on top 4 sensors ===")
print(classification_report(y, y_pred4, target_names=["pass","fail"]))
print("ROC AUC:", round(roc_auc_score(y, scores4), 4))
```

## Cell 18 

```python
# pipeline & serialization for production
# define pipeline: median‐impute IsolationForest
pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("iforest", IsolationForest(
        contamination=0.10,
        random_state=42,
        n_jobs=-1               # use all cores
    ))
])

# fit on your entire dataset (use X_reduced with columns dropped)
# note: X_reduced is before imputation; if you saved it as CSV earlier, reload as needed.
pipe.fit(X_reduced)  

# test its predict/predict_proba interface
# note: IsolationForest uses decision_function; predict returns ±1
flags = pipe.predict(X_reduced)          # +1 = normal, -1 = anomaly
y_pred_prod = (flags == -1).astype(int)
print("Prod performance:\n", classification_report(y, y_pred_prod, target_names=["pass","fail"]))

# serialize the pipeline
os.makedirs("../models", exist_ok=True)
model_path = os.path.join("../models", "secom_iforest_pipeline.joblib")
joblib.dump(pipe, model_path)
print("Saved pipeline to", model_path)
```
